```python
students = {}
def calculate_grade(percentage):
    if percentage >= 75:
        return "A"
    elif percentage >= 60:
        return "B"
    elif percentage >= 50:
        return "C"
    elif percentage >= 35:
        return "D"
    else:
        return "Fail"
        
def add_student():
    roll = input("Enter Roll Number: ")
    name = input("Enter Student Name: ")

    try:
        m1 = int(input("Enter Marks for Subject 1: "))
        m2 = int(input("Enter Marks for Subject 2: "))
        m3 = int(input("Enter Marks for Subject 3: "))
    except ValueError:
        print("‚ùå Invalid marks! Enter numbers only.")
        return

    total = m1 + m2 + m3
    percentage = total / 3
    grade = calculate_grade(percentage)

    students[roll] = {
        "Name": name,
        "Marks": [m1, m2, m3],
        "Total": total,
        "Percentage": percentage,
        "Grade": grade
    }

    print("‚úÖ Student record added successfully!")

def view_student():
    roll = input("Enter Roll Number: ")
    if roll in students:
        s = students[roll]
        print("\n--- Student Result ---")
        print("Name:", s["Name"])
        print("Marks:", s["Marks"])
        print("Total:", s["Total"])
        print("Percentage:", s["Percentage"])
        print("Grade:", s["Grade"])
    else:
        print("‚ùå Student not found!")

def update_student():
    roll = input("Enter Roll Number: ")
    if roll in students:
        try:
            m1 = int(input("Enter New Marks for Subject 1: "))
            m2 = int(input("Enter New Marks for Subject 2: "))
            m3 = int(input("Enter New Marks for Subject 3: "))
        except ValueError:
            print("‚ùå Invalid input!")
            return

        total = m1 + m2 + m3
        percentage = total / 3
        grade = calculate_grade(percentage)

        students[roll]["Marks"] = [m1, m2, m3]
        students[roll]["Total"] = total
        students[roll]["Percentage"] = percentage
        students[roll]["Grade"] = grade

        print("‚úÖ Student record updated!")
    else:
        print("‚ùå Student not found!")

def display_all():
    if not students:
        print("No records available.")
        return

    print("\n--- All Student Records ---")
    for roll, s in students.items():
        print(f"Roll: {roll}, Name: {s['Name']}, Percentage: {s['Percentage']}, Grade: {s['Grade']}")

while True:
    print("\n====== Student Result Management System ======")
    print("1. Add Student")
    print("2. View Student Result")
    print("3. Update Student Marks")
    print("4. Display All Students")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_student()
    elif choice == "3":
        update_student()
    elif choice == "4":
        display_all()
    elif choice == "5":
        print("Thank you! üëã")
        break
    else:
        print("‚ùå Invalid choice! Try again.")
```
